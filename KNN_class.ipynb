```python
import pandas as pd
```

```python
data=pd.read_csv("CKD_data.csv")
```

```python
data=data.drop(['id','Unnamed: 26'],axis=1)
data


```

```python
cat_cols = [col for col in data.columns if data[col].dtype == 'object']
num_cols = [col for col in data.columns if data[col].dtype != 'object']
cat_col
```

```python
data.isnull().sum()
```

```python
data.dropna(0,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['classification'] = data['classification'].map({'ckd':1,'notckd':0})
data['htn'] = data['htn'].map({'yes':1,'no':0})
data['dm'] = data['dm'].map({'yes':1,'no':0})
data['cad'] = data['cad'].map({'yes':1,'no':0})
data['appet'] = data['appet'].map({'good':1,'poor':0})
data['ane'] = data['ane'].map({'yes':1,'no':0})
data['pe'] = data['pe'].map({'yes':1,'no':0})
data['ba'] = data['ba'].map({'present':1,'notpresent':0})
data['pcc'] = data['pcc'].map({'present':1,'notpresent':0})
data['pc'] = data['pc'].map({'abnormal':1,'normal':0})
data['rbc'] = data['rbc'].map({'abnormal':1,'normal':0})
data['wc'] = pd.to_numeric(data['wc'])
data['rc'] = pd.to_numeric(data['rc'])



```

```python
data
```

```python
data.columns
```

```python
#split input and output
independent=data[['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr', 'bu',
       'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad',
       'appet', 'pe', 'ane']]
dependent=data['classification']
```

```python
data['classification'].value_counts()
```

```python
data.info()
```

```python
#split training and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#model selection
from sklearn.neighbors import KNeighborsClassifier
classifier=KNeighborsClassifier()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_pred,y_test)
print(clf_report)
```
